```python
import pandas as pd


chunk_size = 200000

# 1. Identify columns to keep (using the logic we built)
# We'll do a quick pass to get column names and null counts
first_chunk = pd.read_csv(input_file, nrows=1)
all_columns = first_chunk.columns
null_counts = pd.Series(0, index=all_columns)
total_rows = 0

print("Scanning for nulls...")
for chunk in pd.read_csv(input_file, chunksize=chunk_size):
    null_counts += chunk.isnull().sum()
    total_rows += len(chunk)

# Define threshold (e.g., keep columns with less than 50% missing)
limit = total_rows * 0.5
cols_to_keep = null_counts[null_counts <= limit].index.tolist()

# 2. Process and Save to Parquet
print(f"Saving {len(cols_to_keep)} columns to {output_file}...")
first_pass = True
```

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq


chunk_size = 200000

# We use the cols_to_keep list we generated earlier
writer = None

print("Starting the conversion...")

for chunk in pd.read_csv(input_file, usecols=cols_to_keep, chunksize=chunk_size, low_memory=False):

    # 1. Clean the 'No' / Mixed strings in numeric columns
    for col in chunk.select_dtypes(include=['object']).columns:
        if any(word in col for word in ['Factor', 'Area', 'Vol', 'Value', 'Efficiency', 'Loss']):
            chunk[col] = pd.to_numeric(chunk[col], errors='coerce')

    # 2. Convert Pandas chunk to a PyArrow Table
    table = pa.Table.from_pandas(chunk)

    # 3. Initialize the writer on the first chunk
    if writer is None:
        writer = pq.ParquetWriter(output_file, table.schema)

    # 4. Write the chunk
    writer.write_table(table)

# 5. VERY IMPORTANT: Close the writer to finalize the file
if writer:
    writer.close()

print(f"Success! Your file is saved as {output_file}")
```

```python
chunk_size = 200000

writer = None

print("Starting robust conversion...")

for chunk in pd.read_csv(input_file, usecols=cols_to_keep, chunksize=chunk_size, low_memory=False):

    # 1. Force mixed-string columns to numbers (handling the 'No' error)
    for col in chunk.select_dtypes(include=['object']).columns:
        if any(word in col for word in ['Factor', 'Area', 'Vol', 'Value', 'Efficiency', 'Loss', 'Prod', 'Cons']):
            chunk[col] = pd.to_numeric(chunk[col], errors='coerce')

    # 2. SCHEMA FIX: Force all numeric columns to float64
    # This prevents the 'int64' vs 'double' mismatch
    for col in chunk.select_dtypes(include=['number']).columns:
        chunk[col] = chunk[col].astype('float64')

    # 3. Convert to Table
    table = pa.Table.from_pandas(chunk)

    # 4. Initialize or Write
    if writer is None:
        writer = pq.ParquetWriter(output_file, table.schema)

    writer.write_table(table)

if writer:
    writer.close()

print(f"Success! Cleaned data saved to {output_file}")
```

```python
import pandas as pd
import numpy as np

# (Assuming 'df' is your loaded dataframe)
# If you need to reload it, uncomment:
# df = pd.read_csv('/content/drive/MyDrive/BERPublicSearch.txt', sep='\t', on_bad_lines='skip', low_memory=False)

print(f"Initial Shape: {df.shape}")
rows_count = len(df)
threshold = 0.5 * rows_count  # 50% Threshold

print(f"\n--- 🧹 STARTING CLEANUP (Threshold: {int(threshold)} rows) ---")

# Lists to track what we drop and why (for reporting)
cols_dropped_nulls = []
cols_dropped_chars = []
cols_dropped_zeros = []
all_cols_to_drop = set()

# ==========================================
# 1. CHECK NULLS (Standard Missing Data)
# ==========================================
for col in df.columns:
    null_count = df[col].isnull().sum()
    if null_count > threshold:
        cols_dropped_nulls.append(col)
        all_cols_to_drop.add(col)

# ==========================================
# 2. CHECK SUSPICIOUS CHARS (Text Columns)
# ==========================================
text_cols = df.select_dtypes(include=['object']).columns
# We treat '0' and '0.0' as bad data in text columns too
suspicious_chars = ['.', '-', '*', '?', '0', '0.0', 'None', 'nan','none', '?', 'N/A','NAN']

for col in text_cols:
    if col in all_cols_to_drop: continue # Skip if already marked for dropping

    # Count how many rows have these bad characters
    bad_char_count = df[col].astype(str).isin(suspicious_chars).sum()

    # Combined Badness: Nulls + Suspicious Chars
    total_bad = df[col].isnull().sum() + bad_char_count

    if total_bad > threshold:
        cols_dropped_chars.append(col)
        all_cols_to_drop.add(col)

# ==========================================
# 3. CHECK ZEROS (Numeric Columns)
# ==========================================
num_cols = df.select_dtypes(include=['number']).columns

for col in num_cols:
    if col in all_cols_to_drop: continue

    zero_count = (df[col] == 0).sum()

    # Combined Badness: Nulls + Zeros
    total_bad = df[col].isnull().sum() + zero_count

    if total_bad > threshold:
        cols_dropped_zeros.append(col)
        all_cols_to_drop.add(col)

# ==========================================
# 4. EXECUTE DROP
# ==========================================
df_cleaned = df.drop(columns=list(all_cols_to_drop))

# ==========================================
# 5. REPORT
# ==========================================
print(f"\n✅ DROPPED {len(all_cols_to_drop)} COLUMNS TOTAL.")
print(f"   - {len(cols_dropped_nulls)} columns dropped due to Nulls (NaN)")
print(f"   - {len(cols_dropped_chars)} columns dropped due to Bad Chars (., *, etc)")
print(f"   - {len(cols_dropped_zeros)} columns dropped due to >50% Zeros")

print(f"\nNew Shape: {df_cleaned.shape}")

print("\n--- DROPPED COLUMNS ---")
if cols_dropped_nulls: print(f"Nulls: {cols_dropped_nulls} ")
if cols_dropped_chars: print(f"Chars: {cols_dropped_chars} ")
if cols_dropped_zeros: print(f"Zeros: {cols_dropped_zeros} ")
```

```python
import pandas as pd

# Define the output path
output_path = '/content/drive/MyDrive/data_mining_class/BER_Cleaned_Data.parquet'

print(f"Preparing to save {df_cleaned.shape[1]} columns...")

# OPTIONAL: Memory Optimization (Highly recommended for 1.3M rows)
# Converts strings to categories and 64-bit numbers to 32-bit to save RAM
for col in df_cleaned.columns:
    if df_cleaned[col].dtype == 'object':
        df_cleaned[col] = df_cleaned[col].astype('category')
    elif df_cleaned[col].dtype == 'float64':
        df_cleaned[col] = df_cleaned[col].astype('float32')
    elif df_cleaned[col].dtype == 'int64':
        df_cleaned[col] = df_cleaned[col].astype('int32')

# Save to Parquet using snappy compression (balanced speed/size)
df_cleaned.to_parquet(output_path, engine='pyarrow', compression='snappy', index=False)

print(f"✅ FINAL SUCCESS! Your cleaned data is now saved at:\n{output_path}")
```

```python
non_binary_string_columns = []

# Define the set of 'Yes'/'No' values (case-insensitive for robustness)
yes_no_values = {'yes', 'no', 'y', 'n', 'true', 'false', 't', 'f', '1', '0'}

for col in string_columns:
    # Get unique values, convert to lowercase, and drop NaN
    unique_values = df_cleaned[col].dropna().astype(str).str.lower().unique()

    # Check if all unique values are within the 'yes_no_values' set
    # or if the column is entirely empty after dropping NaNs
    if len(unique_values) == 0 or not all(uv in yes_no_values for uv in unique_values):
        non_binary_string_columns.append(col)

print("String columns with data other than just 'Yes'/'No':")
print(non_binary_string_columns)
```

```python
high_cardinality = []
useful_categories = []

total_rows = len(df_cleaned)

for col in df_cleaned.select_dtypes(include=['object', 'category']).columns:
    unique_count = df_cleaned[col].nunique()

    # 1. Very High Cardinality (Likely IDs, names, or addresses)
    if unique_count > (total_rows * 0.5):
        high_cardinality.append((col, unique_count, "Likely ID/Unique"))

    # 2. High Cardinality but potentially useful
    elif unique_count > 15:
        high_cardinality.append((col, unique_count, "High Cardinality Category"))

    # 3. Standard Categories
    else:
        useful_categories.append((col, unique_count))

# Print a nice summary
print(f"{'Column Name':<30} | {'Unique':<10} | {'Type'}")
print("-" * 60)
for col, count, note in high_cardinality:
    print(f"{col:<30} | {count:<10} | {note}")
```

```python
import pandas as pd
import numpy as np

# List of numeric columns from your previous step
numeric_cols = ['Year_of_Construction', 'BerRating', 'GroundFloorArea(sq m)', 'UValueWall',
                'UValueRoof', 'UValueFloor', 'UValueWindow', 'UvalueDoor', 'WallArea',
                'RoofArea', 'FloorArea', 'WindowArea', 'DoorArea', 'NoStoreys', 'CO2Rating',
                'HSMainSystemEfficiency', 'MPCDERValue', 'HSEffAdjFactor', 'HSSupplHeatFraction',
                'HSSupplSystemEff', 'WHMainSystemEff', 'WHEffAdjFactor', 'SupplSHFuel',
                'SupplWHFuel', 'SHRenewableResources', 'WHRenewableResources', 'NoOfFansAndVents',
                'PercentageDraughtStripped', 'NoOfSidesSheltered', 'HeatSystemControlCat',
                'HeatSystemResponseCat', 'NoCentralHeatingPumps', 'DistributionLosses',
                'DeliveredLightingEnergy', 'DeliveredEnergyPumpsFans', 'DeliveredEnergyMainWater',
                'DeliveredEnergyMainSpace', 'PrimaryEnergyLighting', 'PrimaryEnergyPumpsFans',
                'PrimaryEnergyMainWater', 'PrimaryEnergyMainSpace', 'CO2Lighting', 'CO2PumpsFans',
                'CO2MainWater', 'CO2MainSpace', 'GroundFloorArea', 'GroundFloorHeight',
                'FirstFloorArea', 'FirstFloorHeight', 'ThermalBridgingFactor',
                'PredominantRoofTypeArea', 'DeliveredEnergySecondarySpace', 'LivingAreaPercent',
                'CO2SecondarySpace', 'PrimaryEnergySecondarySpace', 'FirstEnergyTypeId',
                'SecondEnergyTypeId', 'ThirdEnergyTypeId', 'FirstWallArea', 'FirstWallUValue',
                'FirstWallAgeBandId', 'FirstWallTypeId', 'CPC', 'EPC']

# 1. Calculate the correlation matrix
corr_matrix = df_cleaned[numeric_cols].corr().abs()

# 2. Extract pairs that are highly correlated
high_corr_pairs = []
threshold = 0.95

# Loop through the upper triangle of the matrix to avoid self-correlation and duplicates
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] >= threshold:
            col_a = corr_matrix.columns[i]
            col_b = corr_matrix.columns[j]
            score = corr_matrix.iloc[i, j]
            high_corr_pairs.append((col_a, col_b, score))

# 3. Display the results
if high_corr_pairs:
    print(f"--- Found {len(high_corr_pairs)} Twin Pairs (Corr >= {threshold}) ---")
    for col_a, col_b, score in high_corr_pairs:
        print(f"🔗 {col_a} <--> {col_b} ({score:.4f})")
else:
    print("No highly redundant pairs found.")
```

```python
# 1. Get the names of the weak features
weak_cols_list = weak_features.index.tolist()

# 2. Combine with your redundancy list
# We use a set to ensure no duplicate names if a column is in both lists
cols_to_remove = list(set(to_drop + weak_cols_list))

print(f"Total columns being pruned: {len(cols_to_remove)}")
print(f"Examples of columns to be removed: {cols_to_remove[:5]}")

# 3. Execute the drop
df_final_weak_non_string = df_final_cleaned.drop(columns=cols_to_remove)

print(f"\n✅ Clean-up Complete!")
print(f"Remaining columns: {df_final_weak_non_string.shape[1]}")
```
