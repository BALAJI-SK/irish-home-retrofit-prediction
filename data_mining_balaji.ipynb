Now, let's identify columns in this new DataFrame (`df_txt`) with more than 50% missing values and columns where more than 50% of the data is just '`.`'.

```python
import pandas as pd

# Path to your downloaded .txt file
input_file = '/content/drive/MyDrive/Documentation/Data_mining_tejas/BERPublicsearch.txt'
# Path where you want to save the new .csv
output_file = '/content/drive/MyDrive/Documentation/Data_mining_tejas/BERPublicsearch.csv'

chunk_size = 50000

first_chunk = True

# Added encoding='cp1252' to handle the special characters
for chunk in pd.read_csv(input_file,
                         sep='\t',
                         chunksize=chunk_size,
                         low_memory=False,
                         encoding='latin-1',
                         on_bad_lines='skip'):

    chunk.to_csv(output_file, mode='a', index=False, header=first_chunk)
    if first_chunk:
        print("Processing first chunk... structure looks good.")
    first_chunk = False

print("Conversion complete!")
```

```python
import pandas as pd

chunk_size = 50000
null_counts = None
total_rows = 0

for chunk in pd.read_csv('/content/drive/MyDrive/Documentation/Data_mining_tejas/BERPublicsearch.csv', chunksize=chunk_size, low_memory=False):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    total_rows += len(chunk)

# Calculate percentages
null_percentages = null_counts / total_rows
columns_to_remove = null_percentages[null_percentages > 0.5].index.tolist()

print("Columns to remove (>50% null):", columns_to_remove)
```

```python
import pandas as pd

input_file = '/content/drive/MyDrive/Documentation/Data_mining_tejas/BERPublicsearch.csv'

# Calculate nulls across chunks
chunk_size = 100000
null_counts = None
total_rows = 0

for chunk in pd.read_csv(input_file, chunksize=chunk_size, low_memory=False, encoding='latin-1'):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    total_rows += len(chunk)

# Get the list
null_percentages = null_counts / total_rows
high_null_cols = null_percentages[null_percentages > 0.5].index.tolist()

print(f"Total Rows: {total_rows}")
print("Columns to remove (>50% null):", high_null_cols)
```
