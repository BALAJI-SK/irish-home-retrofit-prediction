```python
import pandas as pd

file_path = r"C:\Users\achal\Downloads\BER_Final_Cleaned_Data.parquet"
df = pd.read_parquet(file_path)

print("Shape of the data:", df.shape)
df.head()
```

```python

columns_list = df.columns.tolist()
print(columns_list)
```

```python
# Tripathi & Kumar (2024) explicitly state that CO2-related metrics must be removed because they are derived from the Energy Value/Rating. Keeping them leads to a "false" high accuracy.
leakage_columns = [
    'CO2Rating', 
    'CO2Lighting', 
    'CO2PumpsFans', 
    'CO2MainWater', 
    'CO2MainSpace', 
    'CO2SecondarySpace', 
    'MPCDERValue', 
    'CPC'
]

# Drop these columns
df = df.drop(columns=leakage_columns)

# Verify the columns are removed
print(df.columns.tolist())
```

```python
# Usman Ali et al. (2024) used a "refined subset" of 19 variables to achieve 91% accuracy. They argue that redundant data creates "noise" that hides the true physical drivers
redundant_columns = [ 
    'GroundFloorArea(sq m)'
]

# Drop these columns
df = df.drop(columns=redundant_columns)

# Verify columns
print(df.columns.tolist())
```

```python
# Zhang et al. (2023) and Benavente-Peces (2020) explain that non-physical data leads to overfitting, where the model learns the quirks of the dataset rather than the energy behavior of the buildings.
administrative_columns = [
    'TypeofRating', 
    'PurposeOfRating', 
    'MultiDwellingMPRN'
]

# Drop these columns
df = df.drop(columns=administrative_columns)

# Verify columns
print(df.columns.tolist())
```

```python
# Tripathi & Kumar (2024) state that you must remove variables that have a direct mathematical relationship with the target BER rating. If you keep PrimaryEnergyMainSpace, the model will just look at that number to determine the BerRating instead of learning from the UValueWall.
additional_leakage_columns = [
    'EnergyRating', 
    'EPC', 
    'PrimaryEnergyLighting', 
    'PrimaryEnergyPumpsFans', 
    'PrimaryEnergyMainWater', 
    'PrimaryEnergyMainSpace', 
    'PrimaryEnergySecondarySpace'
]

# Drop these columns
df = df.drop(columns=additional_leakage_columns)

# Verify remaining columns
print(df.columns.tolist())
```

```python
# These are outcomes of the building's performance, not the cause. Usman Ali et al. (2024) focus on the 19 input variables (like insulation and system efficiency) because they represent what a user can actually change. Using the "Delivered Energy" as an input is redundant.
calculated_outcome_columns = [
    'DeliveredLightingEnergy', 
    'DeliveredEnergyPumpsFans', 
    'DeliveredEnergyMainWater', 
    'DeliveredEnergyMainSpace', 
    'DeliveredEnergySecondarySpace'
]

# Drop these columns
df = df.drop(columns=calculated_outcome_columns)

# Verify remaining columns
print(df.columns.tolist())
```

```python
# Usman Ali (2024) found that using a single, high-quality representative value for the building envelope leads to a higher accuracy (91%) than splitting it into multiple sub-wall components.
redundant_physical_columns = [
    'FirstWallUValue',
    'FirstWallArea',
    'FirstWallIsSemiExposed',
    'FirstWallAgeBandId',
    'FirstWallTypeId',
]

# Drop these columns
df = df.drop(columns=redundant_physical_columns)

# Verify remaining columns
print(df.columns.tolist())
```

```python
print("Shape of cleaned data:", df.shape)
```

```python
# In the Irish SEAI database, these IDs are often assigned after the calculation engine determines the primary fuel usage. They "leak" the result of the energy efficiency calculation.
data_leakage_columns = [
    'FirstEnergyTypeId',
    'SecondEnergyTypeId',
    'ThirdEnergyTypeId'
]

# Drop the columns
df = df.drop(columns=data_leakage_columns)

# Verify remaining columns
print(df.columns.tolist())
```

```python
# Usman Ali (2024)
multi_collinearity_columns = [
    'GroundFloorArea',     
    'FirstFloorArea',       
    'PredominantRoofTypeArea'  
]

# Drop the columns
df = df.drop(columns=multi_collinearity_columns)

# Verify remaining columns
print(df.columns.tolist())
```

```python
# Samiksha Tripathi (2024) found that many SEAI columns are 90% empty because assessors don't always fill them out
sparse_columns = [
    'SWHPumpSolarPowered',
    'PermeabilityTest',
    'DraftLobby'
]

# Drop the columns
df = df.drop(columns=sparse_columns)

# Verify remaining columns
print(df.columns.tolist())
```

```python
print("Shape of the data:", df.shape)
```

```python
# Zhang (2023) paper suggests that these add more noise than value. Most models perform better with a single FloorArea and NoStoreys.
height_redundancy_columns = [
    'GroundFloorHeight',
    'FirstFloorHeight'
]

# Drop these columns
df = df.drop(columns=height_redundancy_columns)

# Verify remaining columns
print(df.columns.tolist())
```
